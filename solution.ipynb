```python
import os
os.chdir("C:\\Users\\A\\Desktop\\tt")
import pandas as pd
import numpy as np
df = pd.read_csv("iris_data.csv")
df.head()
```

```python
df.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']
df
```

```python
table1 = df['class'].value_counts()
table1
```

```python
table1.set_value("Total",  df.shape[0]).to_frame().reset_index(drop=True)
table1 = table1.to_frame().T
table1
```

```python
def addtocell(df1,df2, index, columns):
    v = [[f"{df1.iloc[i,j]} ({df2.iloc[i,j]})" for j in range(0, df1.shape[1])] for i in range(0,df2.shape[0])]
    return pd.DataFrame(v, index = index,columns = columns)
```

Count is used instead of N

```python
table2 = df.groupby('class')['sepal length'].agg([ 'count',
                                                    'mean', 
                                                    'min',
                                                    'median',
                                                    'max', 
                                                    'std']).T
table2
```

```python
table2['Total'] = [table2.iloc[0].sum(), 
                     table2.iloc[1].mean(),
                     table2.iloc[2].min(), 
                     table2.iloc[3].median(),
                     table2.iloc[4].max(), 
                     table2.iloc[5].std()]
table2
```

```python
table2.index.names = ['SEPAL LENGTH CM']
p = pd.Series([0, 2, 1, 2, 1, 2], table2.index)

table2 = table2.apply(lambda x: x.map(p.map(lambda x: '{{:0.{}f}}'.format(x).format).loc[x.name]), 1)
table2
```

```python
df1 = df.loc[df['sepal length']<5].groupby('class')['sepal length'].agg({"<5":'count'})
df1
```

```python
df2 = df.loc[(df['sepal length']>=5) & (df['sepal length']<6)].groupby('class')['sepal length'].agg({">=5 AND <6":'count'})
df2
```

```python
df3 =df.loc[(df['sepal length']>=6) & (df['sepal length']<7)].groupby('class')['sepal length'].agg({">=6 AND <7":'count'})
df3
```

```python
df4 = df.loc[df['sepal length']>=7].groupby('class')['sepal length'].agg({">=7":'count'})
df4
```

```python
df5 = pd.concat([df1, df2, df3, df4], axis=1).fillna(0).T
df5['Total'] = [df5.iloc[0].sum(), 
                     df5.iloc[1].sum(),
                     df5.iloc[2].sum(), 
                     df5.iloc[3].sum()]
df5 = df5.applymap(np.int64)
df5
```

```python
df6 = df.loc[df['sepal length']<5].groupby('class')['sepal length'].agg({"<5 ":'count'}).apply(lambda x: 100 * x / len(df.index))
df6
```

```python
df7 = df.loc[(df['sepal length']>=5) & (df['sepal length']<6)].groupby('class')['sepal length'].agg({">=5 AND <6 ":'count'}).apply(lambda x: 100 * x / len(df.index))
df7
```

```python
df8 =df.loc[(df['sepal length']>=6) & (df['sepal length']<7)].groupby('class')['sepal length'].agg({">=6 AND <7":'count'}).apply(lambda x: 100 * x / len(df.index))
df8
```

```python
df9 = df.loc[df['sepal length']>=7].groupby('class')['sepal length'].agg({">=7":'count'}).apply(lambda x: 100 * x / len(df.index))
df9
```

```python
df10 = pd.concat([df6,df7,df8,df9], axis=1).fillna(0).T
df10
```

```python
df10['Total'] = [df10.iloc[0].sum(), 
                     df10.iloc[1].sum(),
                     df10.iloc[2].sum(), 
                     df10.iloc[3].sum()]
df10
```

```python
df10 = df10.round(decimals=1)
df10
```

```python
table3 = addtocell(df5, df10, ['<5','>=5 AND <6','>=6 AND <7', '>=7'], ['Iris-setosa','Iris-versicolor','Iris-virginica','Total'])
table3.index.names = ['SEPAL LENGTH (%)']
table3
```
